## Question 3

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.5)
```

## Question 4

```python
reg = LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
pred = reg.predict(X_test)
```

```python
result = pd.DataFrame({"Predicted":pred,"Actual":Y_test})
result
```

```python
reg.score(X_test,Y_test)*100
```

## Question 5

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,pred))
```

```python
from sklearn import model_selection
kfold = model_selection.KFold(n_splits=8,random_state=4)
```

## Classification Accuracy

```python
results = model_selection.cross_val_score(reg,x,y,cv=kfold,scoring='accuracy')
results
```

```python
results.sum()/10
```

## Logarithmic Loss

```python
results = model_selection.cross_val_score(reg,x,y,cv=kfold,scoring='neg_log_loss')
results
```

```python
results.sum()/10
```

## R2 Metrics 

```python
results = model_selection.cross_val_score(reg,x,y,cv=kfold,scoring='r2')
results
```

```python
results.sum()/10
```

```python
from sklearn import metrics
print(metrics.mean_squared_error(Y_test,pred))
```

## Mean Absolute Error

```python
print(metrics.mean_absolute_error(Y_test,pred))
```
